```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# 設定 Selenium 瀏覽器
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 打開 Google Flights 頁面
url = "https://www.google.com/travel/flights/booking?tfs=CBwQAhpIEgoyMDI0LTA5LTI3Ih4KA1RQRRIKMjAyNC0wOS0yNxoDTlJUKgJHSzICMTRqDAgCEggvbS8wZnRreHIMCAISCC9tLzA3ZGZrQAFIAXABggELCP___________wGYAQI&tfu=CmhDalJJWjBSTlUwZDBOa0Z6UmsxQlFWOVVObWRDUnkwdExTMHRMUzB0ZEdKaVptc3lPVUZCUVVGQlIySnplbkl3UVdSUllqaEJFZ1JIU3pFMEdnb0l0U1VRQUJvRFZGZEVPQnh3czNVPRICCAAiAA&authuser=0"
driver.get(url)

# 等待頁面加載
driver.implicitly_wait(10)

# 使用 XPath 抓取包含 NT$5,999 的 div 文字
price_element = driver.find_element(By.XPATH, "//div[contains(@class, 'frOi8 BgYkof fVSoi')]")
price = price_element.text
print(price)

# 使用 XPath 抓取包含 "The least expensive flights for similar trips" 的 div 文字
least_expensive_flight_element = driver.find_element(By.XPATH, "//div[contains(@class, 'NtS4zd eoY5cb')]")
range = least_expensive_flight_element.text
print(range)

# 使用 XPath 抓取帶有 ke9kZe-LkdAo-RbRzK-JNdkSc 和 pKrx3d 類的所有 g 元素
elements = driver.find_elements(By.XPATH, "//*[name()='g' and @class='ke9kZe-LkdAo-RbRzK-JNdkSc pKrx3d']")
data = [element.get_attribute("aria-label") for element in elements]
for entry in data:
    print(entry)

# 匯出到 CSV 檔案
csv_file = "google_flights_data_history.csv"
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Flight Info", "Price History"])  # 寫入表頭
    for entry in data:
        writer.writerow([price, range, entry])  # 每一行寫入一個元素

print(f"資料已匯出到 {csv_file}")

# 關閉瀏覽器
driver.quit()
```
