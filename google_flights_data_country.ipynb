```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import time
import re
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 不顯示瀏覽器窗口（可選）
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 打開 Google Travel 網站
url = 'https://www.google.com/travel/explore?tfs=CBwQAxoqEgoyMDI0LTA5LTE2KABqDAgCEggvbS8wZnRreHIMCAQSCC9tLzAyajcxQAFIAXABggELCP___________wGYAQKyAQQYASAB&tfu=GgA&hl=en&gl=TW'
driver.get(url)

# 等待幾秒鐘以確保頁面加載完成
time.sleep(5)

# 獲取頁面內容
content = driver.page_source

# 在這裡可以使用 BeautifulSoup 來解析頁面內容
soup = bs(content, 'html.parser')

# 找到你想要的元素
# 找h3標籤
h3 = soup.find_all('h3')
for h in h3:
    print(h.text)
    
# 找<span class = 'nx0jzf'標籤 
stops = soup.find_all('span', class_='nx0jzf')
for stop in stops:
    print(stop.text)

# 找<span class = 'Xq1DAb'標籤
times = soup.find_all('span', class_='Xq1DAb')
for time in times:
    print(time.text)
    
# 找<span class = 'QB2Jof xLPuCe'標籤
prices = soup.find_all('span', class_='QB2Jof xLPuCe')
for price in prices:
    print(price.text)

# 等待頁面加載並找到TOKYO的全部航班資料

try:
    city_link = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.XPATH, "//h3[contains(text(), 'Tokyo')]"))
    )
    city_link.click()
    
    # 等待新頁面加載完成
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class='JPLlH sSHqwe']"))
    )

    # 獲取新頁面內容
    city_content = driver.page_source
    city_soup = bs(city_content, 'html.parser')
    
    # 查找所有航班的根元素（每個航班的<div>）
    flights = city_soup.find_all('a', class_='DvoDQ')
    
    for flight in flights:
        # 提取航空公司名稱
        airline_div = flight.find('div', class_='IMgkJe QB2Jof')
        airline_name = airline_div.get_text(strip=True) if airline_div else '未找到航空公司名稱'
        print(f"航空公司名稱: {airline_name}")

        # 提取時間
        time_span = flight.find('span', class_='yApPxd')
        if time_span:
            # 獲取時間文字，包括父元素的直接文字內容
            time_text = time_span.get_text(separator=" ", strip=True)
    
            # 使用正則表達式提取飛行時間 "3 hr 20 min"
            import re
            flight_time_match = re.search(r'\d+ hr \d+ min', time_text)
            flight_time = flight_time_match.group(0) if flight_time_match else '未找到飛行時間'
        else:
            flight_time = '未找到飛行時間'

        # 輸出飛行時間
        print(f"飛行時間: {flight_time}")
    
    
        # 提取出發機場代碼和抵達機場代碼
        airport_codes = flight.find_all('span', {'aria-describedby': True})
        if len(airport_codes) >= 2:
            departure_code = airport_codes[0].get_text(strip=True)
            arrival_code = airport_codes[1].get_text(strip=True)
        else:
            departure_code = '未找到出發機場代碼'
            arrival_code = '未找到抵達機場代碼'
        print(f"出發機場代碼: {departure_code}")
        print(f"抵達機場代碼: {arrival_code}")

        # 提取價格
        price_span = flight.find('span', role='text')
        price = price_span.get_text(strip=True) if price_span else '未找到價格'     
        print(f"價格: {price}")
        print('-' * 40)
        
    
# 關閉瀏覽器
finally:
    driver.quit()
```
